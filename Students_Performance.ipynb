## 7. One-Way ANOVA — GPA by Parental Education Level

> **Hypothesis**  
> H₀: Mean GPA is the same across all parental-education groups.  
> H₁: At least one group has a significantly different mean GPA.  
> Significance level: α = 0.05

```python
from scipy.stats import f_oneway

# Build one GPA array per ParentalEducation level
anova_col = 'ParentalEducation'
groups = [
    df[df[anova_col] == lvl]['GPA'].dropna().values
    for lvl in sorted(df[anova_col].unique())
]

# Report group sizes
print("Group sizes (ParentalEducation level → n):")
for lvl, grp in zip(sorted(df[anova_col].unique()), groups):
    print(f"  Level {lvl}: n = {len(grp)}, mean GPA = {grp.mean():.4f}")

# Run one-way ANOVA
f_stat, p_value = f_oneway(*groups)
print(f"\nF-statistic : {f_stat:.4f}")
print(f"p-value     : {p_value:.6f}")

alpha = 0.05
if p_value < alpha:
    print("\n→ Reject H₀: significant difference in mean GPA across parental-education groups.")
else:
    print("\n→ Fail to reject H₀: no significant difference in mean GPA across groups.")
```

## Axis III — Predictive Modeling & Deep Learning (on this student dataset)

This section applies the requested workflow (dataset preparation, regression, classification, advanced models, deep learning, interpretability, and backup) directly to `df` from this notebook.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

from sklearn.linear_model import LinearRegression, Ridge, Lasso, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import (
    r2_score, mean_squared_error,
    accuracy_score, classification_report, confusion_matrix
)

import joblib
```

```python
# Targets for this dataset
# Regression target: GPA
# Classification target: is_high_performer (1 if GPA >= 3.0 else 0)
df_model = df.copy()
df_model['is_high_performer'] = (df_model['GPA'] >= 3.0).astype(int)

# Build features while avoiding leakage
excluded_cols = ['StudentID', 'GPA', 'GradeClass', 'is_high_performer']
X = df_model.drop(columns=[c for c in excluded_cols if c in df_model.columns], errors='ignore')
y_reg = df_model['GPA']
y_cls = df_model['is_high_performer']

categorical = X.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()
numerical = X.select_dtypes(exclude=['object', 'category', 'bool']).columns.tolist()

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical)
])

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X, y_cls, test_size=0.2, random_state=42, stratify=y_cls
)

print('X shape:', X.shape)
print('Regression target mean:', y_reg.mean().round(3))
print('Classification balance:')
print(y_cls.value_counts(normalize=True).round(3))
```

```python
# 6.2 Regression Models
reg_models = {
    'LinearRegression': LinearRegression(),
    'RandomForestRegressor': RandomForestRegressor(random_state=42),
    'Ridge': Ridge(),
    'Lasso': Lasso()
}

reg_results = []
reg_predictions = {}
reg_pipelines = {}

for name, model in reg_models.items():
    pipe = Pipeline([
        ('preprocess', preprocessor),
        ('regressor', model)
    ])
    pipe.fit(X_train_reg, y_train_reg)
    preds = pipe.predict(X_test_reg)

    reg_pipelines[name] = pipe
    reg_predictions[name] = preds
    reg_results.append({
        'Model': name,
        'R2': r2_score(y_test_reg, preds),
        'MSE': mean_squared_error(y_test_reg, preds)
    })

reg_results_df = pd.DataFrame(reg_results).sort_values('R2', ascending=False)
print(reg_results_df)

best_reg_name = reg_results_df.iloc[0]['Model']
best_reg_preds = reg_predictions[best_reg_name]

plt.figure(figsize=(7, 5))
sns.scatterplot(x=y_test_reg, y=best_reg_preds)
plt.plot([y_test_reg.min(), y_test_reg.max()], [y_test_reg.min(), y_test_reg.max()], 'r--')
plt.xlabel('Actual GPA')
plt.ylabel(f'Predicted GPA ({best_reg_name})')
plt.title(f'Actual vs Predicted GPA - {best_reg_name}')
plt.grid(True)
plt.show()
```

```python
# Cross-validation on linear baseline
pipe_lr = Pipeline([
    ('preprocess', preprocessor),
    ('regressor', LinearRegression())
])

cv_scores = cross_val_score(pipe_lr, X, y_reg, cv=5, scoring='r2')
print('Cross-validated R² scores:', np.round(cv_scores, 4))
print('Mean CV R²:', cv_scores.mean().round(4))
```

```python
# 6.3 Classification Models
cls_models = {
    'LogisticRegression': LogisticRegression(max_iter=1000),
    'SVC': SVC(),
    'KNN': KNeighborsClassifier()
}

cls_results = []
cls_predictions = {}
cls_pipelines = {}

for name, model in cls_models.items():
    pipe = Pipeline([
        ('preprocess', preprocessor),
        ('classifier', model)
    ])
    pipe.fit(X_train_cls, y_train_cls)
    preds = pipe.predict(X_test_cls)

    cls_pipelines[name] = pipe
    cls_predictions[name] = preds

    acc = accuracy_score(y_test_cls, preds)
    cls_results.append({'Model': name, 'Accuracy': acc})

    print(f'\n{name} Accuracy: {acc:.4f}')
    print(classification_report(y_test_cls, preds))

cls_results_df = pd.DataFrame(cls_results).sort_values('Accuracy', ascending=False)
print('\nClassification summary:')
print(cls_results_df)


def plot_confusion(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(title)
    plt.show()

for model_name, model_preds in cls_predictions.items():
    plot_confusion(y_test_cls, model_preds, f'Confusion Matrix - {model_name}')
```

```python
# 6.4 Advanced Methods (RandomForest, XGBoost, LightGBM)
advanced_reg_results = []
advanced_reg_models = {
    'RandomForestRegressor': RandomForestRegressor(random_state=42)
}

try:
    from xgboost import XGBRegressor
    advanced_reg_models['XGBRegressor'] = XGBRegressor(
        random_state=42,
        n_estimators=200,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.9,
        colsample_bytree=0.9
    )
except Exception as e:
    print('XGBoost regressor unavailable:', e)

try:
    from lightgbm import LGBMRegressor
    advanced_reg_models['LGBMRegressor'] = LGBMRegressor(random_state=42)
except Exception as e:
    print('LightGBM regressor unavailable:', e)

for name, model in advanced_reg_models.items():
    pipe = Pipeline([
        ('preprocess', preprocessor),
        ('regressor', model)
    ])
    pipe.fit(X_train_reg, y_train_reg)
    preds = pipe.predict(X_test_reg)
    advanced_reg_results.append({
        'Model': name,
        'R2': r2_score(y_test_reg, preds),
        'MSE': mean_squared_error(y_test_reg, preds)
    })

print(pd.DataFrame(advanced_reg_results).sort_values('R2', ascending=False))
```

```python
advanced_cls_results = []
advanced_cls_predictions = {}
advanced_cls_models = {
    'RandomForestClassifier': RandomForestClassifier(random_state=42)
}

try:
    from xgboost import XGBClassifier
    advanced_cls_models['XGBClassifier'] = XGBClassifier(
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42
    )
except Exception as e:
    print('XGBoost classifier unavailable:', e)

try:
    from lightgbm import LGBMClassifier
    advanced_cls_models['LGBMClassifier'] = LGBMClassifier(random_state=42)
except Exception as e:
    print('LightGBM classifier unavailable:', e)

for name, model in advanced_cls_models.items():
    pipe = Pipeline([
        ('preprocess', preprocessor),
        ('classifier', model)
    ])
    pipe.fit(X_train_cls, y_train_cls)
    preds = pipe.predict(X_test_cls)
    advanced_cls_predictions[name] = preds

    acc = accuracy_score(y_test_cls, preds)
    advanced_cls_results.append({'Model': name, 'Accuracy': acc})

    print(f'\n{name} Accuracy: {acc:.4f}')
    print(classification_report(y_test_cls, preds))

print(pd.DataFrame(advanced_cls_results).sort_values('Accuracy', ascending=False))

for model_name, model_preds in advanced_cls_predictions.items():
    plot_confusion(y_test_cls, model_preds, f'Confusion Matrix - {model_name}')
```

```python
# 6.5 Introduction to Deep Learning (Keras dense networks)
X_all_processed = preprocessor.fit_transform(X)

if hasattr(X_all_processed, 'toarray'):
    X_all_processed = X_all_processed.toarray()

X_train_nn_reg, X_test_nn_reg, y_train_nn_reg, y_test_nn_reg = train_test_split(
    X_all_processed, y_reg, test_size=0.2, random_state=42
)

X_train_nn_cls, X_test_nn_cls, y_train_nn_cls, y_test_nn_cls = train_test_split(
    X_all_processed, y_cls, test_size=0.2, random_state=42, stratify=y_cls
)

print('NN train shape (regression):', X_train_nn_reg.shape)
print('NN train shape (classification):', X_train_nn_cls.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Regression neural network
model_reg = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_nn_reg.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1)
])

model_reg.compile(optimizer='adam', loss='mse', metrics=['mae'])
history_reg = model_reg.fit(
    X_train_nn_reg, y_train_nn_reg,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    verbose=0
)

loss_reg, mae_reg = model_reg.evaluate(X_test_nn_reg, y_test_nn_reg, verbose=0)
preds_nn_reg = model_reg.predict(X_test_nn_reg, verbose=0).ravel()

print(f'NN Regression Test MSE: {loss_reg:.4f}')
print(f'NN Regression Test MAE: {mae_reg:.4f}')
print(f'NN Regression R²: {r2_score(y_test_nn_reg, preds_nn_reg):.4f}')
```

```python
# Classification neural network
model_cls = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_nn_cls.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model_cls.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history_cls = model_cls.fit(
    X_train_nn_cls, y_train_nn_cls,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    verbose=0
)

loss_cls, acc_cls = model_cls.evaluate(X_test_nn_cls, y_test_nn_cls, verbose=0)
probs_nn_cls = model_cls.predict(X_test_nn_cls, verbose=0).ravel()
preds_nn_cls = (probs_nn_cls > 0.5).astype(int)

print(f'NN Classification Test Loss: {loss_cls:.4f}')
print(f'NN Classification Test Accuracy: {acc_cls:.4f}')
print(classification_report(y_test_nn_cls, preds_nn_cls))

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history_cls.history['loss'], label='Train Loss')
plt.plot(history_cls.history['val_loss'], label='Validation Loss')
plt.title('NN Classification Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_cls.history['accuracy'], label='Train Accuracy')
plt.plot(history_cls.history['val_accuracy'], label='Validation Accuracy')
plt.title('NN Classification Accuracy')
plt.legend()
plt.show()

plot_confusion(y_test_nn_cls, preds_nn_cls, 'Confusion Matrix - Neural Network')
```
